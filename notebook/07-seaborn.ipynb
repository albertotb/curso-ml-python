Seaborn es una librería de visualización de alto nivel basada en `matplotlib`. Proporciona una API de alto nivel para realizar gráficos estadísticos.

Tutorial: https://seaborn.pydata.org/tutorial.html    
Galeria de ejemplos: http://seaborn.pydata.org/examples/

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
x = np.random.normal(size=100)
sns.distplot(x)
```

```python
mean, cov = [0, 1], [(1, .5), (.5, 1)]
data = np.random.multivariate_normal(mean, cov, 200)
df = pd.DataFrame(data, columns=["x", "y"])

sns.jointplot(x="x", y="y", data=df)
```

```python
iris = sns.load_dataset("iris")
sns.pairplot(iris)
```

```python
tips = sns.load_dataset("tips")
sns.boxplot(x="day", y="total_bill", hue="time", data=tips)
```

```python
sns.catplot(x="day", y="total_bill", hue="smoker", data=tips, kind="bar")
```

```python
sns.catplot(x="time", y="total_bill", hue="sex", col="day", data=tips, kind="box", height=4, aspect=.5)
```
