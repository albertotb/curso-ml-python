```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
x = np.arange(0, 2*np.pi, 0.1)
y = np.sin(x)
```

```python
y.shape
```

```python
plt.figure()
plt.plot(x, y, 'r')
plt.title('Funcion seno')
plt.xlabel('x')
plt.ylabel('sin(x)')
```

```python
# otra forma de hacer la figura anterior (API orientada a objetos)
fig, axes = plt.subplots()

axes.plot(x, y, 'r')
axes.set_title('Funcion seno')
axes.set_xlabel('x')
axes.set_ylabel('sin(x)')
```

```python
# otra forma de hacer la figura anterior (API orientada a objetos)
fig, axes = plt.subplots()

axes.plot(x, y, 'r', label="seno")
axes.plot(x, np.cos(x), 'b', label="coseno", linewidth=2, linestyle="-.")
axes.set_xlabel('x')
axes.legend();
```

```python
# otra forma de hacer la figura anterior (API orientada a objetos)
fig, axes = plt.subplots()

axes.plot(x, y, 'r', label="seno")
axes.plot(x, np.cos(x), 'b', label="coseno", linewidth=2, linestyle="-.")
axes.set_xlabel('x')
axes.set_xlim([1, 5])
axes.set_ylim([-0.75, 0.75])
axes.legend()
```

```python
# diferencia entre figura y ejes
fig, axes = plt.subplots(1, 2,  figsize=(15, 5))

for i, ax in enumerate(axes):
    ax.plot(x, y**(i+1), 'r')
    ax.set_title('Funcion seno')
    ax.set_xlabel('x')
    ax.set_ylabel('sin(x**{})'.format(i+1))
    
#axes[0][0]
#axes[0][1]
#ESTO NO axes[0, 1]
```

```python
# histogramas
n = np.random.randn(100000)
fig, axes = plt.subplots(1, 2, figsize=(12,4))

axes[0].hist(n)
axes[0].set_title("Histograma")
axes[0].set_xlim((min(n), max(n)))

axes[1].hist(n, cumulative=True, bins=50)
axes[1].set_title("Histograma acumulado")
axes[1].set_xlim((min(n), max(n)));
```

```python
# guardar figuras
fig.savefig("hist.pdf")
```

```python
import numpy as np

n = np.array([0,1,2,3,4,5])
x = np.linspace(0, 5, 10)
xx = np.linspace(-0.75, 1., 100)

fig, axes = plt.subplots(1, 4, figsize=(12,3))

axes[0].scatter(xx, xx + 0.25*np.random.randn(len(xx)))
axes[0].set_title("scatter")

axes[1].step(n, n**2, lw=2)
axes[1].set_title("step")

axes[2].bar(n, n**2, align="center", width=0.5, alpha=0.5)
axes[2].set_title("bar")

axes[3].fill_between(x, x**2, x**3, color="green", alpha=0.5);
axes[3].set_title("fill_between");
```

```python
fig, axes = plt.subplots(2, 2, figsize=(12, 10))

axes[0][0].scatter(xx, xx + 0.25*np.random.randn(len(xx)))
axes[0][0].set_title("scatter")

axes[1][0].step(n, n**2, lw=2)
axes[1][0].set_title("step")

axes[0][1].bar(n, n**2, align="center", width=0.5, alpha=0.5)
axes[0][1].set_title("bar")

axes[1][1].fill_between(x, x**2, x**3, color="green", alpha=0.5);
axes[1][1].set_title("fill_between");
```
